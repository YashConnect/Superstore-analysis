```python
categorical_cols=['Customer_Segment','Category','Region','Payment_Mode','Delivery_Staus','Auto_Reorder']
for col in categorical_cols:
    print(f"----{col}----")
    print(df[col].value_counts())
    plt.figure(figsize=(6,6))
    df[col].value_counts().plot(kind='bar',color='skyblue')
    plt.title(f"Distributtion of {col}")
    plt.xticks(rotation=30)
    plt.tight_layout()
    plt.show()
```

# Numerical Variable Analysis

```python
Numerical_Col=['Quantity','Unit_Price','Discount(%)','Sales_Amount','Cost_Price','Profit','Stock_Left','Reorder_Quantity']
for col in Numerical_Col:
    print(f"----{col}----")
    print(f"Mean:{df[col].mean():.2f}")
    print(f"Median:{df[col].median():.2f}")
    print(f"Std Dev:{df[col].std():.2f}")
    print(f"Min:{df[col].min():.2f}")
    print(f"Max:{df[col].max():.2f}")
    plt.figure(figsize=(8,4))
    plt.subplot(1,2,1)
    df[col].hist(bins=20,color='Lightgreen',alpha=0.7)
    plt.title(f"Distribution of {col}")
    plt.subplot(1,2,2)
    df.boxplot(column=col)
    plt.title(f"Boxplot of {col}")
```

# Bivariate Analysis

```python
print("Sales and Profit by Category")
category_performance=df.groupby(df['Category']).agg({'Sales_Amount':'sum','Profit':'sum','Order_ID':'count'}).rename(columns={'Order_ID':'Order_Count'})
category_performance['Profit Margin']=round(category_performance['Profit']/category_performance['Sales_Amount']*100,2)
print(category_performance)
```

```python
fig,axes=plt.subplots(2,2,figsize=(15,12))
category_performance['Sales_Amount'].plot(kind='bar',ax=axes[0,0],color='blue',alpha=0.7)
axes[0,0].set_title("Total Sales by Category")
axes[0,0].tick_params(axis='x',rotation=30)
category_performance['Profit'].plot(kind='bar',ax=axes[0,1],color='green',alpha=0.7)
axes[0,1].set_title("Total Profit by Category")
axes[0,1].tick_params(axis='x',rotation=30)
category_performance['Profit Margin'].plot(kind='bar',ax=axes[1,0],color='orange',alpha=0.7)
axes[1,0].set_title("Total Profit Margin by Category (%)")
axes[1,0].tick_params(axis='x',rotation=30)
category_performance['Order_Count'].plot(kind='bar',ax=axes[1,1],color='red',alpha=0.7)
axes[1,1].set_title("Total Order by Category")
axes[1,1].tick_params(axis='x',rotation=30)
plt.tight_layout()
plt.show()
```

```python
Regional_Performance=df.groupby(df['Region']).agg({'Sales_Amount':'sum','Profit':'sum','Order_ID':'count'}).rename(columns={'Order_ID':'Order Count'})
Regional_Performance['Profit_Margin(%)']=round((Regional_Performance['Profit']/Regional_Performance['Sales_Amount'])*100,2)
# fig,axes=plt.subplots(2,2,figsize(12,8))
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
Regional_Performance['Profit'].plot(kind='pie')
plt.title('Profit Distribution by region')
plt.subplot(2,2,2)
Regional_Performance['Profit_Margin(%)'].plot(kind='bar')
plt.title("Profit Margin by region")
plt.xticks(rotation=30)
plt.subplot(2,2,3)
Regional_Performance['Sales_Amount'].plot(kind='bar')
plt.title("Sales Distribution by region")
plt.xticks(rotation=30)
plt.subplot(2,2,4)
Regional_Performance['Order Count'].plot(kind='pie',autopct='%1.1f%%')
plt.title("Order Count by region")
plt.tight_layout()
plt.show()
```

```python
df['Order Month']=df['Order_Date'].dt.to_period('M')
df['Order Year']=df['Order_Date'].dt.year
#Montly trends
Monthly_trends=df.groupby(df['Order Month']).agg({'Sales_Amount':'sum','Profit':'sum','Order_ID':'count'}).rename(columns={'Order_ID':'Order Count'})
print("Montly Trend")
plt.figure(figsize=(15,10))
plt.subplot(3,1,1)
Monthly_trends['Order Count'].plot(kind='line',color='blue',marker='o')
plt.title("Montly Order Trend")
plt.ylabel("Total Orders")
plt.grid(True)
plt.subplot(3,1,2)
Monthly_trends['Sales_Amount'].plot(kind='line',color='green',marker='o')
plt.title("Montly Sales Trend")
plt.ylabel("Sale Amount")
plt.grid(True)
plt.subplot(3,1,3)
Monthly_trends['Profit'].plot(kind='line',color='red',marker='o')
plt.title("Montly Profit Trend")
plt.ylabel("Total Profit")
plt.grid(True)
```

```python
print("\n === CORRELATION ANALYSIS === \n")

# Select numerical columns for correlation
numerical_df = df[['Quantity', 'Unit_Price', 'Discount(%)', 'Sales_Amount', 'Cost_Price', 'Profit']]

# Correlation matrix
correlation_matrix = numerical_df.corr()
print("Correlation Matrix:")
print(correlation_matrix, '\n\n')

# Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0,square=True, linewidths=0.5)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

# Customer Segment Analysis

```python
segment_analysis=df.groupby(df['Customer_Segment']).agg({
    'Sales_Amount':['sum','mean'],
    'Profit':['sum','mean'],
    'Order_ID':'count',
    'Discount(%)':'mean'
}).round(2)
segment_analysis.columns=['Total_Sales','Avg_Sales','Total_Profit','Avg_Profit','Total_Orders','Avg_Discount']
print(segment_analysis)
```

```python
fig,axes=plt.subplots(2,2,figsize=(15,10))
segment_analysis['Avg_Discount'].plot(kind='bar',ax=axes[0,0],color='lightblue')
axes[0,0].set_title("The Avg sales By customer segment" )
axes[0,0].tick_params(axis='x',rotation=30)
segment_analysis["Total_Profit"].plot(kind="bar", ax=axes[0, 1], color="lightgreen")
axes[0, 1].set_title("The Total Profit By customer segment")
axes[0, 1].tick_params(axis="x", rotation=30)
segment_analysis["Avg_Profit"].plot(kind="bar", ax=axes[1,0], color="lightyellow")
axes[1,0].set_title("The Avg Planes By customer segment")
axes[1,0].tick_params(axis="x", rotation=30)
segment_analysis["Total_Sales"].plot(kind="bar", ax=axes[1,1], color="orange")
axes[1,1].set_title("The Total By customer segment")
axes[1,1].tick_params(axis="x", rotation=30)
plt.tight_layout()
plt.show()
```

```python
print("\n === KEY INSIGHTS SUMMARY === ")

# Overall Metrics
total_sales = df['Sales_Amount' ].sum()
total_profit = df['Profit' ]. sum()
total_orders = df['Order_ID' ].nunique()
avg_profit_margin = (total_profit / total_sales) * 100

print(f"OVERALL PERFORMANCE:\n")
print(f"Total_Sales: ${total_sales:,.2f}")
print(f"Total_Profit: ${total_profit:,.2f}")
print(f"Total_Orders: {total_orders}")
print(f"Average Profit Margin: {avg_profit_margin :.2f}%")

# Top Performing Categories
top_categories = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
print(f"\nTOP PERFORMING CATEGORIES:\n")
for i, (category, profit) in enumerate(top_categories.items(), 1):
    print(f"{i}. {category}: ${profit:,.2f}")

# Regional Performance
best_region = df.groupby('Region')['Profit'].sum().idxmax()
print(f"\nBEST PERFORMING REGION: {best_region}")

# Delivery Performance
delivery_stats = df["Delivery_Staus"].value_counts(normalize=True) * 100
print(f"\n Delivery Status")
for status,percentage in delivery_stats.items():
    print(f"{status}:{percentage:.1f}%")

# Customer Segments
best_segments=df.groupby('Customer_Segment')['Profit'].sum().idxmax()
print(f"\nMost Profiltable Customer Segment :{best_segments}")
```

```python
Analysis_result={
    'Total_sales':total_sales,
    'Total_orders':total_orders,
    'Total_profit':total_profit,
    'Avg profit margion':avg_profit_margin,
    'Top Categories':top_categories,
    'Best Region':best_region,
    'Best Segment':best_segments
}
Result_df=pd.DataFrame(Analysis_result)
Result_df.to_csv('Eda_Analysis_Summary.csv',index=False)
print(f"\nAnalysis summary saved to 'Eda_Analysis_Summary")
```
